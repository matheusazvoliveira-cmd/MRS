```python
# --- Imports ---
# The notebook uses:
# - geopandas/shapely: load + geometric operations
# - networkx: shortest path on the rail graph
# - scipy.spatial: fast nearest-neighbor snapping (KDTree)
# - ipyleaflet + ipywidgets: interactive map + UI controls

import warnings
warnings.filterwarnings('ignore')

from pathlib import Path
import geopandas as gpd
import pandas as pd
import numpy as np
import networkx as nx

from shapely.geometry import LineString
from shapely.ops import unary_union, snap
from scipy.spatial import cKDTree
from pyproj import Transformer

from ipyleaflet import (
    Map, GeoData, LayersControl, WidgetControl,
    Polyline, Marker, basemaps,
    FullScreenControl
)
import ipywidgets as widgets
```

## 1) Configuration

> ⚠️ CRS rule: only set `SOURCE_CRS_IF_MISSING` if `gdf.crs` prints as **None** after `read_file()`.

### Tips
- `STATION_MAX_DISTANCE_TO_RAIL_M` controls which stations appear in the dropdown.
- `NEAR_STATION_SEARCH_RADIUS_M` controls how far we look around A and B to include extra `CodigoFerr` values.
- `ROUND_M` and `SNAP_TOLERANCE_M` help repair small geometry gaps.


```python
# --- File paths (edit if needed) ---
RAILS_SHP_PATH = Path(r"C:\Users\matheus.deoliveira\OneDrive - Wabtec Corporation\Matheus\wabtec\Advanced Technology\MRS\Malha Ferroviária Federal (shp)\Malha Ferroviária Federal (shp)\Estacoes Code\LInhas\Linhas_BR.shp")
STATIONS_SHP_PATH = Path(r"C:\Users\matheus.deoliveira\OneDrive - Wabtec Corporation\Matheus\wabtec\Advanced Technology\MRS\Malha Ferroviária Federal (shp)\Malha Ferroviária Federal (shp)\Estacoes Code\Estacoes\Estacoes.shp")

# --- Data columns ---
FILTER_COLUMN = "CodigoFerr"   # Rail ownership / operator code
FILTER_VALUE = 11              # MRS code

STATION_NAME_FIELD = "NomeEstaca"  # Station name field
STATION_ID_FIELD = None            # Optional unique station ID field

# --- CRS handling ---
# Only used if CRS is missing in the file
SOURCE_CRS_IF_MISSING = "EPSG:4674"  # set to EPSG:4674 ONLY if gdf.crs is None

# --- UI behavior ---
SHOW_SELECTED_STATIONS = True   # show only A and B markers

# --- Station filtering ---
STATION_MAX_DISTANCE_TO_RAIL_M = 8000

# --- Routing behavior ---
AUTO_INCLUDE_CODIGOFERR_NEAR_SELECTED_STATIONS = True
NEAR_STATION_SEARCH_RADIUS_M = 1500
ROUTE_CODIGOS_DEFAULT = [11]

# --- Topology / connectivity tuning ---
NODE_INTERSECTION = True
ROUND_M = 10
SNAP_ENDPOINTS = True
SNAP_TOLERANCE_M = 25

# --- Warnings / guard rails ---
MAX_SNAP_TO_GRAPH_M = 2000

assert RAILS_SHP_PATH.exists(), f"Rails not found: {RAILS_SHP_PATH}"
assert STATIONS_SHP_PATH.exists(), f"Stations not found: {STATIONS_SHP_PATH}"
```

## 2) Load rails + stations and normalize key fields

```python
# Load shapefiles
rails_all = gpd.read_file(RAILS_SHP_PATH)
stations_all = gpd.read_file(STATIONS_SHP_PATH)

print('Rails CRS (raw):', rails_all.crs)
print('Stations CRS (raw):', stations_all.crs)

# Apply CRS only if missing
if rails_all.crs is None and SOURCE_CRS_IF_MISSING is not None:
    rails_all = rails_all.set_crs(SOURCE_CRS_IF_MISSING)
    print('Rails CRS set to:', SOURCE_CRS_IF_MISSING)
if stations_all.crs is None and SOURCE_CRS_IF_MISSING is not None:
    stations_all = stations_all.set_crs(SOURCE_CRS_IF_MISSING)
    print('Stations CRS set to:', SOURCE_CRS_IF_MISSING)

print('Rails CRS:', rails_all.crs)
print('Stations CRS:', stations_all.crs)

# Normalize CodigoFerr to Int64 (prevents 11 vs 11.0 issues)
rails_all[FILTER_COLUMN] = pd.to_numeric(rails_all[FILTER_COLUMN], errors='coerce').round().astype('Int64')

print('Unique CodigoFerr:', sorted(rails_all[FILTER_COLUMN].dropna().unique().tolist()))

if STATION_NAME_FIELD not in stations_all.columns:
    raise ValueError(f"STATION_NAME_FIELD '{STATION_NAME_FIELD}' not found. Columns: {list(stations_all.columns)}")
```

## 3) Prepare display rails (MRS only) + station dropdown list

```python
# Precompute a metric copy of all rails for routing work (fast reuse)
rails_all_m = rails_all.to_crs(epsg=3857)

# Display = MRS rails only
rails_display = rails_all[rails_all[FILTER_COLUMN] == FILTER_VALUE].copy()
rails_display_ll = rails_display.to_crs(epsg=4326)
rails_display_m = rails_display.to_crs(epsg=3857)

# Stations in meters
stations_m = stations_all.to_crs(epsg=3857)
stations_pts_m = stations_m[stations_m.geometry.type == 'Point'].copy()

# Distance to the MRS network
rail_union_m = rails_display_m.geometry.union_all()
stations_pts_m['dist_to_mrs_m'] = stations_pts_m.geometry.distance(rail_union_m)

stations_near = stations_pts_m[stations_pts_m['dist_to_mrs_m'] <= STATION_MAX_DISTANCE_TO_RAIL_M].copy()
print('Stations near MRS:', len(stations_near), f"(<= {STATION_MAX_DISTANCE_TO_RAIL_M} m)")

# Build station labels shown in dropdown
names = stations_near[STATION_NAME_FIELD].astype(str).fillna('')

if STATION_ID_FIELD and STATION_ID_FIELD in stations_near.columns:
    ids = stations_near[STATION_ID_FIELD].astype(str)
    stations_near['station_label'] = names + ' [' + ids + ']'
else:
    dup = names.duplicated(keep=False)
    stations_near['station_label'] = names
    stations_near.loc[dup, 'station_label'] = names[dup] + ' (idx=' + stations_near.loc[dup].index.astype(str) + ')'

stations_near_ll = stations_near.to_crs(epsg=4326)
stations_lookup = stations_near.set_index('station_label')
labels = sorted(list(stations_lookup.index))

print('Example dropdown labels:', labels[:12])
```

## 4) Routing helpers (graph build + snapping)

This section contains the functions used during **Add Layer**.

Note: To avoid pandas/geopandas `.append()` issues, we always use boolean OR filters (no `.append`).


```python
# CRS transformer for drawing the result (routing uses meters; map uses lat/lon)
to_4326 = Transformer.from_crs('EPSG:3857', 'EPSG:4326', always_xy=True)


def infer_route_codigos(ax, ay, bx, by, radius_m=NEAR_STATION_SEARCH_RADIUS_M):
    # Return CodigoFerr values found near Station A or Station B.
    # This helps when the corridor uses shared/other track segments.

    # Quick bbox crop first
    minx, miny = min(ax, bx) - radius_m, min(ay, by) - radius_m
    maxx, maxy = max(ax, bx) + radius_m, max(ay, by) + radius_m
    subset = rails_all_m.cx[minx:maxx, miny:maxy].copy()

    if len(subset) == 0:
        return []

    a_pt = gpd.points_from_xy([ax], [ay])[0]
    b_pt = gpd.points_from_xy([bx], [by])[0]

    da = subset.distance(a_pt)
    db = subset.distance(b_pt)

    near = subset[(da <= radius_m) | (db <= radius_m)]
    if len(near) == 0:
        return []

    codes = sorted(pd.to_numeric(near[FILTER_COLUMN], errors='coerce')
                   .dropna().round().astype('Int64')
                   .unique().tolist())
    return codes


def build_graph_for_route(ax, ay, bx, by, progress=None, status=None):
    # Build a routing graph for the selected A/B.
    # Returns (G, node_coords, kdtree, node_to_comp, codes)

    def set_prog(val, msg=None):
        if progress is not None:
            progress.value = int(val)
        if status is not None and msg is not None:
            status.value = msg

    set_prog(5, 'Inferring routing codes near stations...')

    # Decide which CodigoFerr values to include
    if AUTO_INCLUDE_CODIGOFERR_NEAR_SELECTED_STATIONS:
        codes = infer_route_codigos(ax, ay, bx, by)
        if FILTER_VALUE not in codes:
            codes = [FILTER_VALUE] + codes
        # De-duplicate
        codes = sorted(list(dict.fromkeys(codes)))
    else:
        codes = ROUTE_CODIGOS_DEFAULT

    set_prog(20, f'Routing codes: {codes} | Cropping rails...')

    # Subset rails by code (EPSG:3857)
    rails_m = rails_all_m[rails_all_m[FILTER_COLUMN].isin(codes)].copy()

    # Crop to bbox around A/B with a margin to speed heavy ops
    margin = 200_000  # 200 km
    minx, miny = min(ax, bx) - margin, min(ay, by) - margin
    maxx, maxy = max(ax, bx) + margin, max(ay, by) + margin
    rails_m = rails_m.cx[minx:maxx, miny:maxy]

    set_prog(35, f'Rail segments in bbox: {len(rails_m)} | Exploding...')

    rails_m = rails_m.explode(index_parts=False).reset_index(drop=True)
    rails_m = rails_m[rails_m.geometry.notnull()]
    rails_m = rails_m[rails_m.geometry.type == 'LineString'].copy()

    if len(rails_m) == 0:
        set_prog(100, 'No rails in the selected area.')
        return nx.Graph(), [], None, {}, codes

    # Optional gap repair
    if SNAP_ENDPOINTS:
        set_prog(45, 'Snapping linework (gap repair)...')
        u = unary_union(list(rails_m.geometry.values))
        rails_m['geometry'] = rails_m.geometry.apply(lambda g: snap(g, u, SNAP_TOLERANCE_M))

    # Optional intersection noding
    if NODE_INTERSECTION:
        set_prog(60, 'Noding at intersections...')
        merged = unary_union(list(rails_m.geometry.values))
        noded = gpd.GeoSeries([merged], crs=rails_m.crs).explode(index_parts=False)
        geoms = list(noded.values)
    else:
        geoms = list(rails_m.geometry.values)

    set_prog(75, 'Building graph nodes/edges...')

    # Node rounding merges near-coincident vertices into the same graph node
    def round_xy(xy, r=ROUND_M):
        return (round(xy[0]/r)*r, round(xy[1]/r)*r)

    node_index = {}
    node_coords = []

    def get_node_id(xy):
        xy = round_xy(xy)
        if xy not in node_index:
            node_index[xy] = len(node_coords)
            node_coords.append(xy)
        return node_index[xy]

    G = nx.Graph()

    # Build edges between consecutive vertices
    for geom in geoms:
        if geom is None or geom.is_empty:
            continue
        coords = list(geom.coords)
        for a, b in zip(coords[:-1], coords[1:]):
            u_id = get_node_id(a)
            v_id = get_node_id(b)
            w = LineString([a, b]).length
            if G.has_edge(u_id, v_id):
                if w < G[u_id][v_id]['weight']:
                    G[u_id][v_id]['weight'] = w
            else:
                G.add_edge(u_id, v_id, weight=w)

    set_prog(85, f'Graph: {G.number_of_nodes()} nodes / {G.number_of_edges()} edges | Building KDTree...')

    node_xy = np.array(node_coords)
    tree = cKDTree(node_xy) if len(node_xy) else None

    # Component map
    comps = list(nx.connected_components(G))
    node_to_comp = {}
    for i, comp in enumerate(comps):
        for n in comp:
            node_to_comp[n] = i

    set_prog(90, 'Graph ready.')
    return G, node_coords, tree, node_to_comp, codes
```

## 5) Interactive map (full screen + collapsible panel + progress)

- The map is larger in the output.
- Click the full screen icon to maximize.
- Expand/collapse the route builder panel.


```python
# Map setup
centroid = rails_display_ll.geometry.union_all().centroid
m = Map(center=(float(centroid.y), float(centroid.x)), zoom=6, basemap=basemaps.CartoDB.Positron)

# Make map bigger
m.layout.height = '720px'
m.layout.width = '100%'

# Full screen control
m.add_control(FullScreenControl(position='topleft'))

# Base rail layer
m.add_layer(GeoData(
    geo_dataframe=rails_display_ll[['geometry']],
    name='MRS Rails (CodigoFerr=11)',
    style={'color': '#2ca02c', 'weight': 3, 'opacity': 0.9}
))

# Layer control (toggle overlays)
m.add_control(LayersControl(position='topright'))

# UI widgets
start_dd = widgets.Dropdown(options=labels, description='Station A:', layout=widgets.Layout(width='420px'))
end_dd   = widgets.Dropdown(options=labels, description='Station B:', layout=widgets.Layout(width='420px'))
color_picker = widgets.ColorPicker(value='#ff7f0e', description='Color:')
layer_name = widgets.Text(value='A→B highlight', description='Layer name:', layout=widgets.Layout(width='420px'))

btn_add = widgets.Button(description='Add Layer', button_style='success')
btn_clear = widgets.Button(description='Clear Highlights', button_style='warning')

# Progress + status
progress = widgets.IntProgress(value=0, min=0, max=100, description='Progress:', bar_style='info', layout=widgets.Layout(width='98%'))
status = widgets.HTML(value='Ready.')

# Log output
out = widgets.Output(layout={'border': '1px solid #444', 'padding': '6px', 'max_height': '220px', 'overflow': 'auto'})

# Collapsible panel (Accordion)
panel_body = widgets.VBox([
    widgets.HBox([start_dd, end_dd]),
    widgets.HBox([color_picker, layer_name]),
    widgets.HBox([btn_add, btn_clear]),
    progress,
    status,
    out
])

accordion = widgets.Accordion(children=[panel_body])
accordion.set_title(0, 'Route builder (click to expand/collapse)')
accordion.selected_index = 0

# Hide/Show toggle
toggle_panel = widgets.ToggleButton(value=False, description='Hide panel', icon='window-minimize')

def on_toggle(_):
    if toggle_panel.value:
        accordion.layout.display = 'none'
        toggle_panel.description = 'Show panel'
        toggle_panel.icon = 'window-restore'
    else:
        accordion.layout.display = 'block'
        toggle_panel.description = 'Hide panel'
        toggle_panel.icon = 'window-minimize'

toggle_panel.observe(on_toggle, names='value')

panel_container = widgets.VBox([toggle_panel, accordion])
m.add_control(WidgetControl(widget=panel_container, position='topleft'))

# State
highlight_layers = []
selected_markers = []


def clear_selected_markers():
    global selected_markers
    for mk in list(selected_markers):
        try:
            m.remove_layer(mk)
        except Exception:
            pass
    selected_markers = []


def show_marker(label):
    if not SHOW_SELECTED_STATIONS:
        return
    row = stations_near_ll[stations_near_ll['station_label'] == label].iloc[0]
    mk = Marker(location=(float(row.geometry.y), float(row.geometry.x)), title=label)
    selected_markers.append(mk)
    m.add_layer(mk)


def on_add_clicked(_):
    with out:
        out.clear_output()
        progress.value = 0
        status.value = '<b>Starting...</b>'

        a_label = start_dd.value
        b_label = end_dd.value

        if not a_label or not b_label:
            status.value = '<span style="color:red"><b>Select Station A and Station B</b></span>'
            return
        if a_label == b_label:
            status.value = '<span style="color:red"><b>Station A and B must be different</b></span>'
            return

        # Station points in meters
        a = stations_lookup.loc[a_label]
        b = stations_lookup.loc[b_label]
        ax, ay = float(a.geometry.x), float(a.geometry.y)
        bx, by = float(b.geometry.x), float(b.geometry.y)

        # Build graph (updates progress & status)
        G, node_coords, tree, node_to_comp, codes = build_graph_for_route(ax, ay, bx, by, progress=progress, status=status)

        print('Routing CodigoFerr used:', codes)
        print('Graph nodes:', G.number_of_nodes(), 'edges:', G.number_of_edges())

        if tree is None or G.number_of_nodes() == 0:
            status.value = '<span style="color:red"><b>Graph empty in this area.</b></span>'
            progress.value = 100
            return

        # Snap A/B to nearest nodes
        status.value = 'Snapping stations to graph...'
        progress.value = 92
        d1, n1 = tree.query([ax, ay], k=1)
        d2, n2 = tree.query([bx, by], k=1)
        n1, n2 = int(n1), int(n2)

        print(f"Snapped A (m): {d1:.1f} | Snapped B (m): {d2:.1f}")

        c1 = node_to_comp.get(n1)
        c2 = node_to_comp.get(n2)
        print('Component A:', c1, '| Component B:', c2)

        if d1 > MAX_SNAP_TO_GRAPH_M or d2 > MAX_SNAP_TO_GRAPH_M:
            print('⚠️ Large snap distance — station may not sit on routed rails')

        if c1 != c2:
            status.value = '<span style="color:red"><b>Disconnected components.</b></span> Increase ROUND_M / SNAP_TOLERANCE_M / NEAR_STATION_SEARCH_RADIUS_M.'
            progress.value = 100
            return

        # Shortest path
        status.value = 'Computing shortest path...'
        progress.value = 95
        try:
            path = nx.shortest_path(G, n1, n2, weight='weight')
        except nx.NetworkXNoPath:
            status.value = '<span style="color:red"><b>No path found.</b></span>'
            progress.value = 100
            return

        # Convert nodes to lat/lon
        status.value = 'Drawing layer...'
        progress.value = 98
        coords = []
        for n in path:
            x, y = node_coords[n]
            lon, lat = to_4326.transform(x, y)
            coords.append((lat, lon))

        nm = (layer_name.value or '').strip() or f"{a_label} → {b_label}"
        poly = Polyline(locations=coords, color=color_picker.value, weight=6, opacity=0.95, name=nm)
        m.add_layer(poly)
        highlight_layers.append(poly)

        clear_selected_markers()
        show_marker(a_label)
        show_marker(b_label)

        status.value = f'<span style="color:green"><b>Added layer:</b></span> {nm} ({len(coords)} vertices)'
        progress.value = 100


def on_clear_clicked(_):
    with out:
        out.clear_output()
        removed = 0
        for lyr in list(highlight_layers):
            try:
                m.remove_layer(lyr)
                removed += 1
            except Exception:
                pass
        highlight_layers.clear()
        clear_selected_markers()
        status.value = f'Removed {removed} highlight layer(s).'
        progress.value = 0
        print('Cleared', removed, 'highlight layer(s).')

btn_add.on_click(on_add_clicked)
btn_clear.on_click(on_clear_clicked)

m
```

## 6) (Optional) Save the last added highlight to GeoJSON

```python
import json
from pathlib import Path

OUT_ROUTE_GEOJSON = Path('last_highlight_route.geojson')

if not highlight_layers:
    print('No highlight layer exists yet. Add one first.')
else:
    last = highlight_layers[-1]
    coords = last.locations  # list of (lat, lon)

    feature = {
        'type': 'Feature',
        'properties': {
            'name': getattr(last, 'name', 'highlight'),
            'color': getattr(last, 'color', '#ff7f0e')
        },
        'geometry': {
            'type': 'LineString',
            'coordinates': [[lon, lat] for (lat, lon) in coords]
        }
    }

    geojson = {'type': 'FeatureCollection', 'features': [feature]}
    OUT_ROUTE_GEOJSON.write_text(json.dumps(geojson, ensure_ascii=False, indent=2), encoding='utf-8')
    print('Saved:', OUT_ROUTE_GEOJSON.resolve())
```
